```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cleaned_data')
```

```python
result = data.groupby(['oxygen_used', 'hired'])[['success','died']].mean().reset_index()
```

```python
result['success']  = result['success'] * 100
```

```python
result['died']  = result['died'] * 100
```

```python
result
```

```python
result.to_csv('Oxygen_result', index=False)
```
