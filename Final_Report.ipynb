```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('Oxygen_result')
data2 = pd.read_csv('Seasonal_result')
data3 = pd.read_csv('Decade_result')
data4 = pd.read_csv('Mountains_result')
```

```python
oxy = data1['oxygen_used'].map({True:'Oxygen used', False:'No oxygen'})
```

```python
hired = data1['hired'].map({True: 'Hired', False:'Not hired'})
```

```python
data1['Category'] = oxy + f" & " + hired
```

```python
data1 = data1.set_index('Category')
```

```python
data1
```

```python
# 3. Select only the math columns and tell Pandas to plot a grouped bar chart
ax = data1[['success', 'died']].plot(kind='bar', figsize=(10,6), color=['green','red'])
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=3)
# 4. Add some nice labels to the chart
plt.title("Success and Death Rates based on Oxygen & Hired Help")
plt.ylabel("Percentage Rate")
plt.xlabel("") # Keeps the bottom clean
plt.xticks(rotation=0) # Keeps the text horizontal so you can read it

# 5. Show the plot
plt.show()
```

```python
data2 = data2.set_index('Season')
```

```python
fig ,axis  = plt.subplots(1,2, figsize=(10,6))
bx = data2[['Death', 'Success', 'Failed', 'Total Climbers']].plot(kind='bar', ax=axis[0] , color=['red','green','yellow', 'blue'])
axis[0].set_title("Success and Death based on Seasons")
axis[0].set_ylabel("No of Climbers")
axis[0].set_xlabel("") # Keeps the bottom clean
axis[0].tick_params(axis='x', rotation=0)

by = data2[['Death Rate', 'Success Rate', 'Failed Rate']].plot(kind='bar', ax=axis[1] , color=['red','green','yellow', 'blue'])
axis[1].set_title("Success and Death Rate based on Seasons")
axis[1].set_ylabel("Percenage of Rates")
axis[1].set_xlabel("") # Keeps the bottom clean
axis[1].tick_params(axis='x', rotation=0)

for container in by.containers:
    by.bar_label(container, fmt='%.1f%%', padding=3)

plt.tight_layout()
plt.show()
```
