```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing cleaned data
data = pd.read_csv('cleaned_data')
```

```python
#removing Not-required columns/data
data = data.drop( columns =['peak_name', 'year', 'hired' , 'age', 'oxygen_used'])
```

```python
#filling missing seasons data with "Spring" season
data.loc[data['season'] == 'Unknown', 'season'] = 'Spring'
```

```python
#getting No of seasons
seasons = data['season'].value_counts().reset_index()
```

```python
#creating  DataFrame/final table as "Result" and first column with "Seasons" name
Result  = pd.DataFrame(columns=["Season"])
```

```python
# alloting no of seasons list to seasons column in Result table
Result['Season'] =seasons['season']
```

```python
# Taking data from season,sucsess,died column from data table and computing climbers who died in middle and
# also climbers who died after success and adding there count 
# to get total No of deaths while climbing  
                
                #computing climbers who got success but also died for each season
death_column =  [(len(data[(data['season'] == s) & (data['success'] == 1) & (data['died'] == 1)])) +
                 
                #computing climbers who didnt get success and died for each season
                 (len(data[(data['season'] == s) & (data['died'] == 1)& (data['success'] == 0)])) for s in Result['Season']]
```

```python
Result['Death'] = death_column  #creating death column in Result and assingning it death values for each season
```

```python
# Taking data from season,sucsess,died column from data table and computing climbers who succeded for each season and giving the seasonal count
success_column = [len(data[(data['season'] == s) & (data['success'] == 1) & (data['died'] == 0)]) for s in Result['Season']]
```

```python
# creating sucess column in Result and assingning it success  values for each season
Result['Success'] = success_column
```

```python
# Taking data from season,sucsess,died column from data table and computing climbers who failed for each season and giving the seasonal count
failed_column =  [len(data[(data['season'] == s) & (data['success'] == 0) & (data['died'] == 0)]) for s in Result['Season']]
```

```python
# creating failed column in Result and assingning it failed values for each season
Result['Failed'] = failed_column
```

```python
# computing total No of Climbers each season
climbers_column = Result['Death'] + Result['Success'] + Result['Failed']
```

```python
# assigning Total no of climbers column to rusult table
Result['Total Climbers'] = climbers_column
```

```python
# computing death rate each season
death_rate_column = (Result['Death'] / Result['Total Climbers'] *100).round(2)

# assigning death rate column to result
Result['Death Rate'] = death_rate_column  
```

```python
# computing success rate each season
success_rate_column = (Result['Success'] / Result['Total Climbers'] *100).round(2)

# assigning success rate column to result
Result['Success Rate'] = success_rate_column  
```

```python
 # computing failed rate each season
Failed_rate_column = (Result['Failed'] / Result['Total Climbers'] *100).round(2)

# assigning failed rate column to result
Result['Failed Rate'] = Failed_rate_column
```
