```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing cleaned data
data = pd.read_csv('cleaned_data')
```

```python
#removing Not-required columns/data
data = data.drop(columns= ['peak_name', 'season', 'age', 'hired','oxygen_used','Unnamed: 0']) 
```

```python
# 1. Create a new column using the floor division math trick
data['decade'] = (data['year'] // 10) * 10
```

```python
# 2. (Optional) If you want it to look like "1980s" instead of "1980" for a cleaner table
data['Decade'] = data['decade'].astype(str) + 's'
```

```python
# 3. Group by your new decade column to see the total climbers per decade
Result = data.groupby(['Decade']).size().reset_index(name='Total Climbers')
```

```python
# View the result
Result
```

```python
Result.to_csv('Decade_result', index=False)
```
